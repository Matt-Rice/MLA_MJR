## 9. Write the code to add a column named temp_f to the dataset that contains the temperature from temp_c represented in Fahrenheit.

```python
# Method to convert c to f
def c_to_f(temp_c):
  return ((9/5) * temp_c) + 32

# Adding a column that converts temp c to temp f
df['temp_f'] = c_to_f(df['temp_c'])

print(df.head())
```

## 10. Write the code to rearrange the columns in the dataset so temp_f appears immediately before temp_c.

```python
# Rearranging the columns
df = df.reindex(['date_time', 'temp_f', 'temp_c', 'dew_point', 'relative_humidity', 'wind_speed', 'visibility', 'pressure', 'weather'], axis=1)

print(df.head())
```

## 11. Write the code to display the data types of all the columns in the dataset.

```python
# Printing datatypes
print(df.dtypes)
```

## 12. The date_time column is not actually considered a date by Pandas.  Write the code to convert this column to a date/time column and to verify the conversion was successful.

```python
# Converting date_time column to datetime type
df['date_time'] = pd.to_datetime(df['date_time'])

print(df.dtypes)
```

## 13. Write the code to get the average temperature (in F) for each month and write it into an object called monthly_avg. This object should be a Pandas Series which is similar to a Pandas Dataframe except it only has one column.  You will plot this data later.

```python
# Creating a list of the months
months = df['date_time'].dt.month

# Grouping the means of temp_f by month
monthly_avg = df.groupby(months)['temp_f'].mean()

print(monthly_avg)
```

## 14. Write code to generate a scatterplot that compares the temperature against the relative humidity.  The size of your plot should be 20 x 10.

```python
# Creating Scatterplot with x axis being temperature and y axis being relative humidity
plt.figure(figsize = (20, 10))

sct_plot = sns.scatterplot(x = df['temp_f'], y = df['relative_humidity'])
sct_plot.set(xlabel = 'Temperature (F)', ylabel = 'Relative Humidity')

plt.title('Temperature vs Relative Humidity')
plt.show()
```

## 15. Write the code to generate a line chart that shows the average temperature per month. The size of your plot should 20 x 10.

```python
# Creating PD series of months
month_series = pd.Series(months.unique())

# Generate a line plot with avg temperature per month as the y axis and the month as the x axis
plt.figure(figsize = (20, 10))
ln_plot = sns.lineplot(x = month_series, y = monthly_avg)
ln_plot.set(xlabel = 'Months', ylabel = 'Average Temperature (F)')
plt.title('Average Temperature Per Month')

plt.show()
```
